After reviewing the data, we will eliminate any information that is not essential to our problem statement and keep just what is necessary. The "LandAndOceanAverageTemperature" column displays the average temperature of the land and seas and this overall represents more meaningful temperature readings.

The global temperature dataset's monthly degree of detail. Once a year, it may be resampled to collect fewer entries while maintaining the average temperature information.

```python
global_temperature_df_lo = global_temperatures_df[['dt', 'LandAndOceanAverageTemperature']]
# Check the number of missing values and the corresponding years
print(global_temperature_df_lo.isnull().sum())
print(global_temperature_df_lo[global_temperature_df_lo['LandAndOceanAverageTemperature'].isnull()].index)
```

Dropping Null Values

```python
global_temperature_df_lo.dropna(inplace = True)
print(global_temperature_df_lo.isnull().sum()) 
```
